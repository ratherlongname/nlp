```python
from googletrans import Translator

translator = Translator()
# txt = translator.translate("Hello" , dest="hi")
# print(txt)

with open('title_list.txt') as f:
    mylist = f.read().splitlines()

listcur = mylist[0:1]
for title in listcur:
    pos_name = 'segmented/positive/' + title + '_pos_hi_reviews.txt'
    neg_name = 'segmented/negative/' + title + '_neg_hi_reviews.txt'
    pos_file = open(pos_name, 'r')
    neg_file = open(neg_name, 'r')
    pos_list = pos_file.read().splitlines()
    for line in pos_list:
        print(line, end=' -->  ')
        txt = translator.translate(line , dest="hi")
        print(txt.text)
```
